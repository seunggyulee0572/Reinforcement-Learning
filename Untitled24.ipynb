midi 파일에서 beat, pitch, velocity 정보 추출을 위한 전처리 함수 및 64 beat를 뽑기 위한 슬라이딩 함수

```python
from typing import List, Tuple
from collections import defaultdict

def extract_chords_by_beat(
    midi_path: str
) -> Tuple[float, List[float], List[List[int]], List[List[int]]]:
    """
    한 MIDI 파일에서 beat(박) 단위로 화음 + velocity 정보를 추출.

    return:
        bpm: 추정된 BPM
        beat_times: 각 beat 시작 시각(초)
        chords_by_beat: beat별 unique pitch 리스트 (정렬)
        velocities_by_beat: beat별 각 pitch의 대표 velocity (chords_by_beat와 같은 길이/순서)
    """
    pm = pretty_midi.PrettyMIDI(midi_path)

    # (1) 템포
    tempo_changes, tempi = pm.get_tempo_changes()
    bpm = float(tempi[0]) if len(tempi) > 0 else 120.0

    # (2) beat 시점
    beat_times = pm.get_beats()
    if len(beat_times) < 2:
        return bpm, beat_times, [], []

    # (3) 노트 수집
    notes = []
    for inst in pm.instruments:
        notes.extend(inst.notes)

    chords_by_beat: List[List[int]] = []
    velocities_by_beat: List[List[int]] = []

    approx_beat_len = beat_times[1] - beat_times[0]

    for i in range(len(beat_times)):
        t_start = beat_times[i]
        if i < len(beat_times) - 1:
            t_end = beat_times[i + 1]
        else:
            t_end = t_start + approx_beat_len

        # 이 beat 구간에 걸쳐 있는 노트 (pitch, vel)
        active_pairs = [
            (n.pitch, n.velocity) for n in notes
            if (n.start < t_end) and (n.end > t_start)
        ]

        # pitch별 velocity 모으기
        vel_map = defaultdict(list)
        for p, v in active_pairs:
            vel_map[p].append(v)

        # pitch 정렬 + 대표 velocity (여기선 평균)
        pitches = []
        vels = []
        for p in sorted(vel_map.keys()):
            v_list = vel_map[p]
            rep_v = int(sum(v_list) / len(v_list))  # 평균
            pitches.append(p)
            vels.append(rep_v)

        chords_by_beat.append(pitches)
        velocities_by_beat.append(vels)

    return bpm, beat_times, chords_by_beat, velocities_by_beat

def make_64beat_windows(
    chords_by_beat: List[List[int]],
    velocities_by_beat: List[List[int]],
    window_size: int = 64
) -> List[Tuple[List[List[int]], List[List[int]]]]:
    """
    beat 단위 화음/velocity 타임라인에서
    길이 64짜리 윈도우(chords_64, vels_64)들을 슬라이딩 윈도우로 뽑는다.
    """
    windows = []
    total_beats = len(chords_by_beat)
    if total_beats < window_size:
        return windows  # 64 beat보다 짧으면 스킵

    for start in range(0, total_beats - window_size + 1):
        chords_64 = chords_by_beat[start : start + window_size]
        vels_64   = velocities_by_beat[start : start + window_size]
        windows.append((chords_64, vels_64))

    return windows
```

```python
import os

root_dir = "/content/midi_sample_300"

chord_windows = []   # env에 바로 넣을 리스트
data_dict = {}       # 원하면 파일 단위 정보도 따로 저장

for file_idx, (cur_dir, subdirs, files) in enumerate(os.walk(root_dir)):
    for fname in files:
        if not fname.lower().endswith(".mid"):
            continue
        midi_path = os.path.join(cur_dir, fname)

        try :
        # 1) beat 단위 화음 + velocity 추출
          bpm, beat_times, chords_by_beat, velocities_by_beat = extract_chords_by_beat(midi_path)

          # beat 수가 너무 작으면 패스
          if len(chords_by_beat) == 0:
              continue

          # 2) 64 beat 윈도우들 뽑기
          windows = make_64beat_windows(chords_by_beat, velocities_by_beat, window_size=64)

          # 전역 chord_windows 리스트에 추가
          chord_windows.extend(windows)

          # 디버깅용/분석용으로 파일 단위 정보도 저장
          data_dict[midi_path] = {
              "bpm": bpm,
              "beat_times": beat_times,
              "chords_by_beat": chords_by_beat,
              "velocities_by_beat": velocities_by_beat,
              "n_windows_64": len(windows),
          }
        except Exception as e:
          print(e)

print("환경에 쓸 64-beat 윈도우 개수:", len(chord_windows))
```

멜로디 환경 구성



```python
import numpy as np
import random
from typing import List, Tuple

class MelodyEnv:
    def __init__(
        self,
        chord_windows: List[Tuple[List[List[int]], List[List[int]]]],
        pitch_vocab: List[int],
        vel_bins: List[int],
        max_steps: int = 64,
        history_len: int = 8,
    ):
        """
        chord_windows: [(chords_by_beat_64, vels_by_beat_64), ...]
                       각 요소는 길이 64인 beat별 화음/velocity 정보
        pitch_vocab: 사용할 pitch 값 리스트 (예: [31..96] 또는 스케일 기반)
        vel_bins: 사용할 velocity 값 리스트 (예: [32, 64, 96, 127])
        max_steps: 에피소드 길이 (64 step = 16마디)
        history_len: state에 포함할 최근 멜로디 길이
        """
        self.chord_windows = chord_windows
        self.pitch_vocab = pitch_vocab
        self.vel_bins = vel_bins
        self.max_steps = max_steps
        self.history_len = history_len
        self.state_dim = None
        # action space 크기 (pitch x vel)
        self.n_pitch = len(pitch_vocab)
        self.n_vel = len(vel_bins)
        self.n_actions = self.n_pitch * self.n_vel

        # 에피소드별 상태
        self.episode_chords = None  # 길이 64, 각 원소: [p1, p2, ...]
        self.episode_vels = None    # 길이 64, 각 원소: [v1, v2, ...]
        self.t = 0
        self.melody_history = []

    def reset(self, seed=None):
      # seed가 주어지면 random, numpy seed 고정
      if seed is not None:
          random.seed(seed)
          np.random.seed(seed)
          # deterministic index 선택
          idx = seed % len(self.chord_windows)
          self.episode_chords, self.episode_vels = self.chord_windows[idx]
      else:
          # 기존처럼 랜덤 선택
          self.episode_chords, self.episode_vels = random.choice(self.chord_windows)

      # 초기화
      self.t = 0
      self.melody_history = []

      # 첫 state 생성
      state = self._build_state()

      # state dimension 기록
      if self.state_dim is None:
          self.state_dim = state.shape[0]

      return state

    def encode_action(self, pitch: int, vel: int) -> int:
      # pitch, vel 값이 각각 몇 번째 인덱스인지 찾고
      pitch_idx = self.pitch_vocab.index(pitch)
      vel_idx   = vel

      # 디코딩에서 하던 반대 연산
      action_idx = pitch_idx * self.n_vel + vel_idx
      return action_idx

    def _decode_action(self, action_idx: int) -> Tuple[int, int]:
        """
        action_idx (0 ~ n_actions-1) 를 (pitch, vel)로 변환
        총 pitch(127) * vel(8) = 1024개의 action
        action 에서 pitch와 vel을 분리
        """
        pitch_idx = action_idx // self.n_vel
        vel_idx = action_idx % self.n_vel

        pitch = self.pitch_vocab[pitch_idx]
        vel = self.vel_bins[vel_idx]
        return pitch, vel

    def _reward(self, pitch: int, vel: int) -> float:
      """
      현재 time step(self.t)에서의 보상 계산
      """
      reward = 0.0
      current_chord = self.episode_chords[self.t]      # [p1, p2, ...]
      current_chord_vels = self.episode_vels[self.t]   # [v1, v2, ...]  있다고 가정

      # 1) 코드와의 어울림 (기본 화성 보상)
      if pitch in current_chord:
          reward += 1.0
      else:
          reward -= 0.5

      # 2) 이전 멜로디와의 관계 (반복 패널티)
      last_pitch = None
      last_vel = None
      if self.melody_history:
          last_pitch, last_vel = self.melody_history[-1]
          if last_pitch == pitch:
              reward -= 0.3  # 같은 음 반복 패널티

      # 3) velocity 관련 규칙
      #    - 코드 전체의 평균 velocity와 너무 동떨어지지 않게
      if current_chord_vels:
          avg_chord_vel = sum(current_chord_vels) / len(current_chord_vels)
          diff = abs(vel - avg_chord_vel)

          # 코드 다이내믹과 잘 맞는 경우
          if diff < 15:
              reward += 0.3   # 코드와 비슷한 세기 → 자연스러움
          elif diff < 35:
              reward += 0.1   # 좀 차이나지만 허용 가능한 범위
          else:
              reward -= 0.2   # 너무 튀는 세기 → 살짝 패널티


      # 예: 직전보다 조금 세게 / 조금 약하게는 +, 완전 극단 점프는 -
      if last_vel is not None:
          vel_change = vel - last_vel
          # 너무 과한 변화는 살짝 패널티
          if abs(vel_change) > 60:
              reward -= 0.2
          # 약간의 변화(표현)는 살짝 보상
          elif 10 <= abs(vel_change) <= 40:
              reward += 0.1

      # 4) 음정 이동(interval) 관련 규칙
      if last_pitch is not None:
          interval = abs(pitch - last_pitch)

          if interval == 0:
              pass
          elif interval <= 2:
              # 반음~온음 정도의 작은 움직임 → 선율적으로 안정적
              reward += 0.3
          elif interval <= 5:
              # 3~4도 정도 → 자연스러운 진행
              reward += 0.1
          elif interval == 6:
              # 트라이톤(증4/감5) 특유의 긴장감
              reward -= 0.1
          elif interval >= 12:
              # 한 옥타브 이상 큰 점프 → 자주 나오면 부자연스럽다고 가정
              reward -= 0.4

      return reward

    def _build_state(self) -> np.ndarray:
        """
        state 구성:
        - 최근 history_len 개의 (pitch, vel) (없으면 0으로 패딩)
        - 현재 코드 정보
        [history_pitch, history_vel, current_chord_padded] 단순 벡터로 구성
        """
        # 최근 멜로디 history_len개
        hist = self.melody_history[-self.history_len:]
        pad_len = self.history_len - len(hist)
        # (pitch, vel) -> (2,)
        hist_array = np.zeros((self.history_len, 2), dtype=np.float32)
        for i, (p, v) in enumerate(hist, start=pad_len):
            hist_array[i] = [p, v]

        # 현재 코드 (최대 k개까지만, 나머지는 0 패딩)
        current_chord = self.episode_chords[self.t]
        max_chord_notes = 8  # 예: 코드 최대 8음까지만 사용
        chord_vec = np.zeros((max_chord_notes,), dtype=np.float32)
        for i, p in enumerate(current_chord[:max_chord_notes]):
            chord_vec[i] = p

        # flatten 해서 하나의 state 벡터로
        state_vec = np.concatenate([hist_array.flatten(), chord_vec], axis=0)
        return state_vec

    def step(self, action_idx: int):
        pitch, vel = self._decode_action(action_idx)

        reward = 0.0
        # 보상 계산
        reward = self._reward(pitch, vel)

        # 멜로디 기록
        self.melody_history.append((pitch, vel))

        # 다음 step 준비
        self.t += 1
        done = (self.t >= self.max_steps)

        if not done:
            next_state = self._build_state()
        else:
            # 마지막 step이면 state는 의미 없으니 0으로 리턴해도 됨
            next_state = np.zeros(self.state_dim, dtype=np.float32)

        info = {}
        return next_state, reward, done, info
```

DQN 및 ReplayBuffer 클래스 구현

```python
import torch
import torch.nn as nn
import torch.optim as optim

class DQN(nn.Module):
    def __init__(self, state_dim: int, n_actions: int):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Linear(256, n_actions),
        )

    def forward(self, x):
        return self.net(x)

import random
from collections import deque
import numpy as np
class ReplayBuffer:
    def __init__(self, capacity: int):
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        # state / next_state 는 np.ndarray
        self.buffer.append((
            state.astype(np.float32),
            action,
            float(reward),
            next_state.astype(np.float32),
            bool(done)
        ))

    def sample(self, batch_size: int):
        batch = random.sample(self.buffer, batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)
        return (
            np.stack(states),
            np.array(actions, dtype=np.int64),
            np.array(rewards, dtype=np.float32),
            np.stack(next_states),
            np.array(dones, dtype=np.float32),
        )

    def __len__(self):
        return len(self.buffer)
```

모델을 돌리기 위한 하이퍼파라미터 세팅

```python
def get_epsilon(step: int):
    if step >= eps_decay_steps:
        return eps_end
    # 선형 감소
    return eps_start - (eps_start - eps_end) * (step / eps_decay_steps)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 네트워크
policy_net = DQN(state_dim, n_actions).to(device)
target_net = DQN(state_dim, n_actions).to(device)
target_net.load_state_dict(policy_net.state_dict())
target_net.eval()

optimizer = optim.Adam(policy_net.parameters(), lr=1e-4)

# 리플레이 버퍼
replay_buffer = ReplayBuffer(capacity=50_000)

# DQN 관련 설정
batch_size = 64
gamma = 0.5
target_update_interval = 1000   # step마다 타겟 네트워크 동기화
min_buffer_size = 1000          # 이만큼 쌓이기 전까지는 학습 안 함
max_episodes = 5000             # 에피소드 수 (원하는 만큼)
max_steps_per_episode = env.max_steps  # 64

# 입실론-그리디 탐험
eps_start = 1.0
eps_end = 0.05
eps_decay_steps = 50_000  # 5만 step 동안 선형 감소
global_step = 0
```

실제 모델 학습

```python
import math
from tqdm import trange

episode_rewards = []

for episode in trange(max_episodes):
    state = env.reset()
    episode_reward = 0.0

    for t in range(max_steps_per_episode):
        global_step += 1

        # --- epsilon-greedy 액션 선택 ---
        eps = get_epsilon(global_step)
        if random.random() < eps:
            # 탐험
            action = random.randrange(n_actions)
        else:
            # exploitation: Q값 최대인 액션
            with torch.no_grad():
                s_tensor = torch.from_numpy(state).unsqueeze(0).to(device)  # (1, state_dim)
                q_values = policy_net(s_tensor)  # (1, n_actions)
                action = int(torch.argmax(q_values, dim=1).item())

        # --- 환경 한 스텝 진행 ---
        next_state, reward, done, info = env.step(action)
        episode_reward += reward

        # 리플레이 버퍼에 저장
        replay_buffer.push(state, action, reward, next_state, done)
        state = next_state

        # --- 학습 (버퍼 크기 충분할 때만) ---
        if len(replay_buffer) >= min_buffer_size:
            states_b, actions_b, rewards_b, next_states_b, dones_b = replay_buffer.sample(batch_size)

            states_t      = torch.from_numpy(states_b).to(device)           # (B, state_dim)
            actions_t     = torch.from_numpy(actions_b).unsqueeze(1).to(device)  # (B, 1)
            rewards_t     = torch.from_numpy(rewards_b).unsqueeze(1).to(device)  # (B, 1)
            next_states_t = torch.from_numpy(next_states_b).to(device)      # (B, state_dim)
            dones_t       = torch.from_numpy(dones_b).unsqueeze(1).to(device)    # (B, 1)

            # Q(s,a)
            q_values = policy_net(states_t).gather(1, actions_t)  # (B,1)

            # max_a' Q_target(s', a')
            with torch.no_grad():
                next_q_values = target_net(next_states_t).max(dim=1, keepdim=True)[0]  # (B,1)
                target = rewards_t + gamma * (1.0 - dones_t) * next_q_values

            loss = nn.MSELoss()(q_values, target)

            optimizer.zero_grad()
            loss.backward()
            # gradient clipping (선택)
            nn.utils.clip_grad_norm_(policy_net.parameters(), max_norm=1.0)
            optimizer.step()

            # 타겟 네트워크 주기적으로 동기화
            if global_step % target_update_interval == 0:
                target_net.load_state_dict(policy_net.state_dict())

        if done:
            break

    episode_rewards.append(episode_reward)

    if (episode + 1) % 50 == 0:
        avg_last = np.mean(episode_rewards[-50:])
        print(f"[Episode {episode+1}] 최근 50개 평균 reward: {avg_last:.3f}, epsilon: {eps:.3f}")
```

```python
def build_pitch_vocab(chord_windows):
    """
    chord_windows를 기반으로 pitch_vocab 생성.
    데이터에 등장한 pitch들을 unique + 정렬해서 리스트로 반환.
    """
    pitch_set = set()
    for chords_64, _ in chord_windows:
        for beat_pitches in chords_64:
            for p in beat_pitches:
                pitch_set.add(p)

    pitch_vocab = sorted(pitch_set)
    return pitch_vocab
```

```python
import math
from tqdm import trange

episode_rewards = []
epsilon_history = []

for episode in trange(max_episodes):
    state = env.reset()
    episode_reward = 0.0

    for t in range(max_steps_per_episode):
        global_step += 1

        # --- epsilon-greedy 액션 선택 ---
        eps = get_epsilon(global_step)
        if random.random() < eps:
            # 탐험
            action = random.randrange(n_actions)
        else:
            # exploitation: Q값 최대인 액션
            with torch.no_grad():
                s_tensor = torch.from_numpy(state).unsqueeze(0).to(device)  # (1, state_dim)
                q_values = policy_net(s_tensor)  # (1, n_actions)
                action = int(torch.argmax(q_values, dim=1).item())

        # --- 환경 한 스텝 진행 ---
        next_state, reward, done, info = env.step(action)
        episode_reward += reward

        # 리플레이 버퍼에 저장
        replay_buffer.push(state, action, reward, next_state, done)
        state = next_state

        # --- 학습 (버퍼 크기 충분할 때만) ---
        if len(replay_buffer) >= min_buffer_size:
            states_b, actions_b, rewards_b, next_states_b, dones_b = replay_buffer.sample(batch_size)

            states_t      = torch.from_numpy(states_b).to(device)           # (B, state_dim)
            actions_t     = torch.from_numpy(actions_b).unsqueeze(1).to(device)  # (B, 1)
            rewards_t     = torch.from_numpy(rewards_b).unsqueeze(1).to(device)  # (B, 1)
            next_states_t = torch.from_numpy(next_states_b).to(device)      # (B, state_dim)
            dones_t       = torch.from_numpy(dones_b).unsqueeze(1).to(device)    # (B, 1)

            # Q(s,a)
            q_values = policy_net(states_t).gather(1, actions_t)  # (B,1)

            # max_a' Q_target(s', a')
            with torch.no_grad():
                next_q_values = target_net(next_states_t).max(dim=1, keepdim=True)[0]  # (B,1)
                target = rewards_t + gamma * (1.0 - dones_t) * next_q_values

            loss = nn.MSELoss()(q_values, target)

            optimizer.zero_grad()
            loss.backward()
            # gradient clipping (선택)
            nn.utils.clip_grad_norm_(policy_net.parameters(), max_norm=1.0)
            optimizer.step()

            # 타겟 네트워크 주기적으로 동기화
            if global_step % target_update_interval == 0:
                target_net.load_state_dict(policy_net.state_dict())

        if done:
            break
    epsilon_history.append(eps)

    episode_rewards.append(episode_reward)

    if (episode + 1) % 50 == 0:
        avg_last = np.mean(episode_rewards[-50:])
        print(f"[Episode {episode+1}] 최근 50개 평균 reward: {avg_last:.3f}, epsilon: {eps:.3f}")
```

평가를 위한 함수

```python
import numpy as np
import random
import copy
from typing import Callable, Dict, Any

import torch

# 타입: policy_fn(state, env, t) -> action_idx
PolicyFn = Callable[[np.ndarray, "MelodyEnv", int], int]


# ========================================
# 1. 에피소드 실행 + Metric 계산
# ========================================
def run_episode_and_collect_metrics(env: "MelodyEnv",
                                    policy_fn: PolicyFn) -> Dict[str, Any]:
    """
    한 에피소드를 돌리면서:
      - 총 reward (return)
      - 코드톤 안에 들어간 음 비율
      - 평균 도약 크기 (세미톤)
      - pitch 분산
    을 계산해서 반환.
    """
    state = env.reset()
    done = False
    total_reward = 0.0

    pitches = []          # 멜로디 pitch 시퀀스
    chord_hit_count = 0   # 코드 안에 들어간 음 개수
    t = 0                 # step index
    chords = env.episode_chords  # reset() 이후 이번 에피소드의 코드 진행

    while not done:
        action_idx = policy_fn(state, env, t)
        next_state, reward, done, info = env.step(action_idx)

        total_reward += reward

        # action_idx -> pitch 디코딩 (env._decode_action 대신 직접 계산)
        pitch_idx = action_idx // env.n_vel
        pitch = env.pitch_vocab[pitch_idx]
        pitches.append(pitch)

        # 현재 step의 코드톤 안에 들어가는지 체크
        current_chord = chords[t]
        if pitch in current_chord:
            chord_hit_count += 1

        state = next_state
        t += 1

    # ----- 음악 이론 기반 Metric 계산 -----
    if len(pitches) >= 2:
        leaps = [abs(pitches[i] - pitches[i - 1]) for i in range(1, len(pitches))]
        avg_leap = float(np.mean(leaps))
    else:
        avg_leap = 0.0

    pitch_variance = float(np.var(pitches)) if pitches else 0.0
    ratio_in_chord = float(chord_hit_count / len(pitches)) if pitches else 0.0

    return {
        "total_reward": float(total_reward),
        "avg_leap": avg_leap,
        "pitch_variance": pitch_variance,
        "ratio_in_chord": ratio_in_chord,
        "n_steps": len(pitches),
    }


def evaluate_policy(env: "MelodyEnv",
                    policy_fn: PolicyFn,
                    n_episodes: int = 50,
                    ma_window: int = 50) -> Dict[str, Any]:
    """
    주어진 policy를 여러 에피소드에 대해 평가하고,
    - 에피소드별 return 리스트
    - 평균 return
    - 최근 ma_window개 에피소드 평균 return
    - 음악 이론 기반 Metric들의 에피소드 평균
    을 반환.
    """
    episode_metrics = []
    returns = []

    for ep in range(n_episodes):
        m = run_episode_and_collect_metrics(env, policy_fn)
        episode_metrics.append(m)
        returns.append(m["total_reward"])

    returns = np.array(returns, dtype=np.float32)

    # RL 관점 Metric
    avg_return = float(np.mean(returns))
    if len(returns) >= ma_window:
        recent_avg_return = float(np.mean(returns[-ma_window:]))
    else:
        recent_avg_return = float(np.mean(returns))

    # 음악 이론 기반 Metric 평균
    avg_ratio_in_chord = float(np.mean([m["ratio_in_chord"] for m in episode_metrics]))
    avg_leap = float(np.mean([m["avg_leap"] for m in episode_metrics]))
    avg_pitch_var = float(np.mean([m["pitch_variance"] for m in episode_metrics]))

    return {
        "returns": returns,                        # 에피소드별 return
        "avg_return": avg_return,                 # 전체 평균 return
        "recent_avg_return": recent_avg_return,   # 최근 ma_window개의 평균
        "avg_ratio_in_chord": avg_ratio_in_chord, # 코드톤 비율 평균
        "avg_leap": avg_leap,                     # 평균 도약 크기 평균
        "avg_pitch_variance": avg_pitch_var,      # pitch 분산 평균
    }


# ========================================
# 2. Baseline Policies
# ========================================
def random_policy(state: np.ndarray, env: "MelodyEnv", t: int) -> int:
    """
    완전 랜덤 정책:
    action space 전체에서 균등 샘플링.
    """
    return random.randrange(env.n_actions)


# ========================================
# 3. DQN policy (학습된 policy_net 사용)
# ========================================
def dqn_greedy_policy(state: np.ndarray, env: "MelodyEnv", t: int) -> int:
    """
    학습된 policy_net을 사용해서
    항상 argmax Q(s,a)를 선택하는 greedy DQN 정책.
    """
    policy_net.eval()
    state_tensor = torch.from_numpy(state).float().unsqueeze(0).to(device)  # (1, state_dim)
    with torch.no_grad():
        q_values = policy_net(state_tensor)  # (1, n_actions)
        action_idx = int(q_values.argmax(dim=1).item())
    return action_idx



# ========================================
# 4. 에피소드 1개 플레이 + MIDI 저장
# ========================================
def play_episode_and_save_midi(env: "MelodyEnv",
                               policy_fn: PolicyFn,
                               midi_path: str,
                               seed: int = 123) -> float:
    """
    주어진 policy_fn으로 에피소드 1개를 실행하고,
    - 총 reward를 반환
    - 해당 에피소드의 코드, velocity, 멜로디를 MIDI로 저장
    """
    # env를 복사해서 원래 env 상태에 영향 안 주도록
    local_env = copy.deepcopy(env)

    state = local_env.reset(seed=seed)
    done = False
    total_reward = 0.0
    t = 0

    while not done:
        action_idx = policy_fn(state, local_env, t)
        state, reward, done, _ = local_env.step(action_idx)
        total_reward += reward
        t += 1

    episode_chords = local_env.episode_chords
    episode_vels = local_env.episode_vels
    melody = local_env.melody_history

    midi_path = create_midi_from_episode(
        episode_chords,
        episode_vels,
        melody,
        midi_path
    )

    print(f"[MIDI 저장 완료] path={midi_path}, total_reward={total_reward:.3f}")
    return total_reward


# ========================================
# 5. 사용 예시
# ========================================

# env, policy_net, device, n_actions, create_midi_from_episode
# 가 이미 준비되어 있다고 가정

# 1) 정책별 성능 평가 (여러 에피소드 기준)
random_stats = evaluate_policy(env, random_policy,
                               n_episodes=100,
                               ma_window=50)
print("Random policy:", random_stats)


dqn_stats = evaluate_policy(env, dqn_greedy_policy,
                            n_episodes=100,
                            ma_window=50)
print("DQN greedy policy:", dqn_stats)

# 2) 동일한 seed로 1개 에피소드 생성 + MIDI 비교

# 2-1) 랜덤 정책
_ = play_episode_and_save_midi(
    env,
    random_policy,              # 완전 랜덤 정책
    "/content/random.mid",
    seed=123
)

# 2-2) 학습된 DQN 정책
_ = play_episode_and_save_midi(
    env,
    dqn_greedy_policy,          # DQN으로 학습된 정책
    "/content/policy.mid",
    seed=123
)

```
